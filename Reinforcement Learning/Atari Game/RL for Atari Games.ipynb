```python
environment_name='Breakout-v0'
env=gym.make(environment_name)
```

```python
env.reset()
```

```python
env.action_space
```

```python
env.observation_space
```

```python
episodes = 5
for episode in range(1, episodes+1):
    state = env.reset()
    done = False
    score = 0 
    
    while not done:
        env.render()
        action = env.action_space.sample()
        n_state, reward, done, info = env.step(action)
        score+=reward
    print('Episode:{} Score:{}'.format(episode, score))
env.close()
```

```python
env.close()
```

## Vectorised environment and train model

```python
env=make_atari_env('Breakout-v0',n_envs=4,seed=0)                 #creating 4 environments using n_envs=4
env=VecFrameStack(env,n_stack=4)                                  #stacking those 4 environments together
```

```python
env.reset()
```

```python
env.render()
```

```python
env.close()
```

```python
log_path=os.path.join('Training','Logs')
model=A2C('CnnPolicy',env,verbose=1,tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=100000)
```

```python
## Save and reload
```

```python
a2c_path=os.path.join('Training','Saved Models', 'A2C_Breakout_Model')
model.save(a2c_path)
```

```python
del model
```

```python
model=A2C.load(a2c_path,env)
```

## Evaluate and test

```python
#we can evaluate only one model
env=make_atari_env('Breakout-v0',n_envs=1,seed=0)                
env=VecFrameStack(env,n_stack=4)   
```

```python
evaluate_policy(model,env,n_eval_episodes=10,render=True)
```
